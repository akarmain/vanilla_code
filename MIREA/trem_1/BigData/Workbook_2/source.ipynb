### 1.3.1. Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами

```python
import numpy as np

chess = np.zeros((8, 8), dtype=int)
chess[1::2, ::2] = 1
chess[::2, 1::2] = 1
print(chess)
```

### 2.3.1. Найти евклидово расстояние между двумя Series

```python
import pandas as pd
import math

a = pd.Series([1, 2, 3])
b = pd.Series([4, 6, 8])

sum_sq = 0
for x, y in zip(a, b):
    sum_sq += (x - y) ** 2

distance = math.sqrt(sum_sq)
print("Series a:")
print(a)
print("\nSeries b:")
print(b)
print("\nЕвклидово расстояние:", distance)
```

### 3.3.2. Нормализация признаков набора iris

Для выполнения этой ячейки нужен доступ к интернету, так как набор данных загружается по ссылке из задания.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"

r = requests.get(url, verify=False)
data = StringIO(r.text)
iris = pd.read_csv(data)

minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

iris["sepal_length_cm_minmax"] = minmax_scaler.fit_transform(iris[["sepal_length_cm"]])
iris["sepal_width_cm_zscore"] = standard_scaler.fit_transform(iris[["sepal_width_cm"]])

print(iris[[
    "sepal_length_cm",
    "sepal_length_cm_minmax",
    "sepal_width_cm",
    "sepal_width_cm_zscore"
]].head(10))
```
